### Tipos de experimentos

- Experimento determinístico

  São aqueles em que o resultado é previsível.

  Exemplo: Temperatura do ar.
  

- Experimento aleatório
  
  São aqueles em que o resultado não pode ser previsto.

  Exemplo: Experimento de um jogo de dados é um experimento aleatório.



### Experimento aleatório

É aquele experimento que, quando repetido em iguais condições, pode fornecer resultados diferentes, ou seja, são resultados explicados ao acaso, isto
é, não podem ser previamente determinados, dependem exclusivamente
do acaso.

Características do experimento aleatório:
   
   - O experimento pode ser repetido.
   - Embora não seja possível afirmar que resultado em particular ocorrerá, é possível descrever o conjunto de todos os resultados possíveis do experimento.
   - À medida que aumenta o número de repetições aparece uma certa regularidade que torna possível a construção de um modelo matemático.

## Espaço amostral

É o conjunto universo ou o conjunto de todos os resultados possíveis de um experimento aleatório. A letra que representa o espaço amostral é S.

Exemplo:

Num jogo de futebol, entre duas equipes, uma das equipes pode obter
resultados tais como: vitória (v), empate (e) ou derrota (d). Tem-se então:
S = {v, e, d}. Portanto: n (S) = 3

## Evento

É um conjunto qualquer de resultados de um experimento aleatório. Pode-se dizer que um evento é qualquer subconjunto de um espaço amostral. É representado, pela letra E.

Exemplo 

No lançamento de duas moedas, o espaço amostral é E = {(C,C), (C,K), (K,C),
(K,K)}. Se aparecer faces iguais, o conjunto C = {(K,K), (C,C)}, subconjunto de
E, é um evento de E.

### Tipos de eventos

- Eventos certos – são eventos que possuem todos os elementos do
espaço amostral. Logo, se E = S, ele é o próprio espaço amostral.

- Eventos impossíveis – são eventos que não possuem elementos no espaço
amostral. Logo, se E = Ø, o evento E é chamado de evento impossível.

- Eventos elementares – se E ⊂ S (E está contido em S) e E é um conjunto
unitário, então E é chamado evento elementar.

- Evento complementar – em relação a determinado evento A, podemos definir seu evento complementar (A) que é caracterizado pela não ocorrência daquele evento.

## Conceito de probabilidade

Se em um fenômeno aleatório, as possibilidades de ocorrência são igualmente provaveis, então a probabilidade de ocorrência de um evento A é igual ao número de casos favoráveis de ocorrência de A dividido pelo número total de casos possíveis ou espaço amostral. 

### Exemplo

Fazendo-se inspeção em um lote de 240 peças de motor, o departamento
de controle de qualidade constatou que 20 peças estavam com defeito.
Retirando-se ao acaso uma das 240 peças, a probabilidade de esta peça
NÃO ser defeituosa é:

```python
## exemplo

### espaço amostral
S_n = 240

### número de peças defeituosas
n_defeituosas = 20
### número de peças não defeituosas
n_sem_defeitos = S_n - n_defeituosas

### probabilidade de peça não defeituosa
P_semdefeitos = round ((n_sem_defeitos/S_n),3)*100
print("A chance sair uma peça não defeituosa é de ",P_semdefeitos, '%')
 

```

```python
def min_multiplo_comum(a,b):
    j = 2
    c = np.array([])
    mmc = np.array([])
    while True: 
        if (a % j == 0 and b%j==0):
            c = np.append(c,np.array([a/j]))
            c = np.append(c,np.array([b/j]))
            mmc = np.append(mmc,np.array([j]))
            #j = j + 1
            a = a/j
            b = b/j
            if (a % j != 0 and b%j!=0):
                j = j + 1
            
        elif (a % j == 0 and b%j!=0):
            c = np.append(c,np.array([a/j]))
            a = a/j
            mmc = np.append(mmc,np.array([j]))
            if (a % j != 0):
                j = j + 1
        elif (a % j != 0 and b%j==0):
            c = np.append(c,np.array([b/j]))
            b = b/j
            mmc = np.append(mmc,np.array([j]))
            if (b % j != 0):
                j = j + 1
            #j = j + 1
        else:
            break
    return [np.prod(mmc),c]
           
```

### Funções Comuns de Probabilidade

```python
random.seed(1)
np.random.seed(1)
random_state = 1

## Permutação, possibilidades de colocar n valores em uma posição. (factorial de n)
def permutacao(n):
    return math.factorial(n)

## P objetos em n posições, a ordem importa. (equação: !n/(n-p)!
def arranjo(n,p):
    return math.factorial(n)/math.factorial(n-p)

## Combinação de n objetos em p posições, a ordem não importa. (equação: !n/p!(n-p)!
## Cn^p
def combinacao(n,p):
    return math.factorial(n)/(math.factorial(p)*math.factorial(n-p))

## possibilidades possiveis de n slots, em p possibilidades para cadaa um.
def possibilidades(n,p):
    return p**n
```

Distribuições discretas
Bernoulli - distribuição que representa uma escolha binária, com probabilidades p e 1-p. Como exemplo um lançamento de uma moeda.

Binomial - representa a soma de sucessos em n execuções de um processo binário. Como exemplo o número de caras em n lançamentos de uma moeda. Outro exemplo é o número de bolas pretas retiradas de um cesto contendo bolas brancas e pretas (com reposição).

Hipergeométrica - representa a soma de bolas pretas retiradas de um cesto contendo bolas brancas e pretas (sem reposição).

Poisson - número de chamdas recebidas num intervalo de tempo.

Discrete Uniform - cada ponto do espaço tem igual probabilidade. Como exemplo o lançamento de um dado.

Geométrica - número de fracassos antes de um sucesso. Por exemplo, no lançamento de moeda, o número de coroas antes de uma cara.

Binomial negativa - número de fracassos antes de n sucessos



### Distribuição Binomial

A densidade binomial é uma das mais importantes em teoria da probabilidade.
Ela surge como uma idealização matemática de diversas situações comuns
na “vida real” e está intimamente ligada à amostragem com reposição.

A situação clássica em que usamos uma densidade binomial é a seguinte.
Uma experiência tem apenas 2 resultados possíveis : “sucesso” e “falha”, em que a probabilidade de “sucesso” é p e a probabilidade de “falha” é q = 1 − p .


A experiência é repetida um número fixo (n) de vezes, sempre nas mesmas condições, de tal forma que as probabilidades de “sucesso” (p) e “falha” ( q = 1 − p ) se mantêm inalteradas a cada repetição. As diversas repetições da experiência são feitas de maneira independente, ou seja, o resultado de
uma repetição não afeta o resultado das outras.

A variável aleatória X que mede o número de “sucessos” nas n repetições da experiência é uma variável discreta, com valores possíveis 0, 1, 2, .... ,n. Dizemos que esta variável tem densidade binomial com parâmetros n e p, e escrevemos que X ~ Bin(n, p).

```python
## Exemplo distribuição binomial.
## O cruzeiro , no brasileirão  tem 4/5 de probabilidade de vitória sempre que  joga dentro de casa. 
## Se o cruzeiro jogar 20 partidas em casa calcule a probabilidade de vencer exatamente 15 partidas.
## Equação:  P(x=k) = combinacao(n,k)*p^k*q^n-k

k=15
n=20
p=4/5
q=1-p

P = combinacao(n,k)*(p**k)*(q**(n-k))*100
print(P) 

## Usando biblioteca scipy
rv = stats.binom(n,p)
p1 = rv.pmf()
print(p1)
```

### Distribuição Hipergeométrica

- Usadas para calcular probabilidades no caso de amostragem sem reposição.

Equação: 

P(X=k)= combinacao(r,x)*combinacao((n-x),(N-r))/combinacao(n,N)

onde:
 - N é o número total de elementos
 - r é o número de elementos que satisfazem a condição
 - x é o número x que se quer calcular a probabilidade.
 - n é o subconjunto de elementos que estão sendo considerados.


```python
## Exemplo 1:
## Uma caixa contém 6 bolas azuis, 4 brancas. Numa amostra de 5 bolas , qual a probabilidade de 3 serem azuis?

N = 10
r = 6
n = 5
x = 3

P = (combinacao(r,x)*combinacao(N-r,n-x))/combinacao(N,n)
print(P)

```

```python
## Exemplo 2
## Uma caixa contém 12 lampadas, das quais 5 estão queimadas . São escolhidas  6 lâmpadas.
## a) Qual a probabilidade de que nenhuma esteja queimada.
## b) Qual a probabilidade de que pelo menos uma esteja queimada.

N = 12
r = 5 #  queimadas
n = 6 # subconjunto de lampadas
x = 0 # nenhuma queimada

P = (combinacao(r,x)*combinacao(N-r,n-x))/combinacao(N,n)
print(P)
```

### Distribuição Poisson

Esta distribuição é usada principalmente para modelar o número de ocorrências de um evento “raro” (de probabilidade baixa) durante um intervalo de tempo especificado.

Equação:

 P(x) = (e^-λ) * (λ^x)/x!
 onde x = ocorrencias
     λ = média ou taxa de ocorrencias

```python
## Exemplo
## Considere que ocorrem , em média 2 mortes por colera ao mês num batalhão.
## a) Qual a probabilidade de não haver mortes em 1 mês ?
## b) qual a probabilidade de haver 1 morte em um mês?
## c) qual a probabilidade de haver 3 mortes em 2 meses?
## d) qual a probabilidade de haver 3 mortes em 3 meses?

λ1 = 2 # taxa de morte 
x1 = 0 #número de mortes
P1 = (math.e**(-λ1)*(λ1**x1))/math.factorial(x1) * 100  
print(P1)

λ2 = 2 
x2 = 1
P2 = (math.e**(-λ2)*(λ2**x2))/math.factorial(x2) * 100  
print(P2)

λ3 = 2*2 #porque a taxa é ao mês, porém estamos analisando dois mêses 
x3 = 3
P3 = (math.e**(-λ3)*(λ3**x3))/math.factorial(x3) * 100  
print(P3)

λ4 = 2*3 #porque a taxa é ao mês, porém estamos analisando três mêses 
x4 = 3
P4 = (math.e**(-λ4)*(λ4**x4))/math.factorial(x4) * 100  
print(P4)

```
